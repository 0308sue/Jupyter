```python
from haversine import haversine
import warnings
warnings.filterwarnings(action='ignore')
import numpy as np
```

```python
import pandas as pd
df = pd.read_excel('가축전염병(가축3종)_최근3년(발생일자별 날씨)_위경도수정.xlsx')
df.head()
```

```python
# country 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
chicken = df['축종(품종)'] == '닭'

# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
dfc = df[chicken]

# 결과를 출력합니다.
dfc

dfcc = dfc.iloc[:,0:7]
dfccc = dfc.iloc[:,7:8]
# dfcccc = dfc.iloc['축종(품종)']
dfccccc = dfc.iloc[:,14:17]
dfcccccc = dfc.iloc[:,19:23]
dfcc = pd.concat([dfccc,dfcc,dfccccc,dfcccccc],axis=1)
dfcc.head()
```

```python
dfa = pd.read_excel('경기도경상북도_겨울철새서식개체수201710202203 (2).xlsx')
dfa
dfaa = dfa.iloc[:,1:5]
dfaaa = dfa[['해당년월']]
dfaa = pd.concat([dfaaa,dfaa],axis=1)
dfaa.head()
```

```python
result = pd.DataFrame(columns=['발생일자(가)','가축전염병명(가)','광역시도(가)','시군구(가)','읍면동(가)','리(가)','위도(가)','경도(가)','관측지코드','관측지일자코드','평균기온','일교차','일강수량','평균풍속','최다풍향','해당년월(서)','시도(서)','시군(서)','위도(서)','경도(서)'])
listc = []
aa = 0
for c in range(len(dfcc)):
    dfccr = dfcc.loc[c]
    la = float(dfccr['위도'])
    lo = float(dfccr['경도'])
    chicken = (la,lo)
    cc = c+1
    for a in range(len(dfaa)):
        dfaar = dfaa.loc[a]
        laa = float(dfaar['위도'])
        loa = float(dfaar['경도'])
        ai = (laa,loa)
#         # 거리 계산
        listc.append(haversine(chicken, ai, unit = 'km'))
        dfcdic = dfccr.to_list()
        dfadic = dfaar.to_list()
        dfdf = dfcdic+dfadic
        aa = aa+1
        bb = aa*cc
        i=bb-1
        result.loc[i] = dfdf
result
```

```python
result['거리(KM)'] = listc
result
```
